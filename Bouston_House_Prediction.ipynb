```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('BostonHousing.csv')
```

```python
import seaborn as sns 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
df
```

```python
# to check for outliers in 'medv' - cloumn
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.distplot(df['medv'], bins=30)
plt.show()
```

```python
sns.boxplot(df['medv'])
```

```python
# Step 1: Calculate Q1, Q3, and IQR
Q1 = df['medv'].quantile(0.25)
Q3 = df['medv'].quantile(0.75)
IQR = Q3 - Q1

# Step 2: Compute Minimum and Maximum Limits for Outliers
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

print(f"Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")
print(f"Lower Limit: {lower_limit}, Upper Limit: {upper_limit}")

# Step 3 & 4: Identify and Replace Outliers
df.loc[df['medv'] < lower_limit, 'medv'] = Q1
df.loc[df['medv'] > upper_limit, 'medv'] = Q3

print(df)
```

```python
sns.boxplot(df['medv'])
```

```python
df.corr()
```

```python
correlation_matrix = df.corr().round(2)
# annot = True to print the values inside the square
# plt.figure(figsize = (16,9))
sns.heatmap(data=correlation_matrix, annot=True, cmap='RdPu',linewidths=.5, fmt=".2f", cbar_kws={"shrink": 0.75})
# plt.show()
```

```python
plt.figure(figsize=(20, 5))

features = ['lstat', 'rm']
target = df['medv']

for i, col in enumerate(features):
    plt.subplot(1, len(features) , i+1)
    x = df[col]
    y = target
    plt.scatter(x, y, marker='o')
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('MEDV')
```

```python
X = pd.DataFrame(np.c_[df['lstat'], df['rm']], columns = ['lstat','rm'])
Y = df['medv']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)
```

```python
# model evaluation for training set
y_train_predict = lin_model.predict(X_train)
rmse = (np.sqrt(mean_squared_error(Y_train, y_train_predict)))
r2 = r2_score(Y_train, y_train_predict)

print("The model performance for training set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")

```

```python
# model evaluation for testing set
y_test_predict = lin_model.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))
r2 = r2_score(Y_test, y_test_predict)

print("The model performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```

```python
# Scatter plot for training set
plt.figure(figsize=(12, 6))
plt.scatter(Y_train, y_train_predict, c='blue', label='Actual vs. Predicted (Training)')
plt.plot([min(Y_train), max(Y_train)], [min(Y_train), max(Y_train)], linestyle='--', color='red', linewidth=2, label='Ideal Line')
plt.title('Linear Regression - Training Set')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.legend()
plt.show()

# Predictions for testing set
y_test_predict = lin_model.predict(X_test)

# Scatter plot for testing set
plt.figure(figsize=(12, 6))
plt.scatter(Y_test, y_test_predict, c='green', label='Actual vs. Predicted (Testing)')
plt.plot([min(Y_test), max(Y_test)], [min(Y_test), max(Y_test)], linestyle='--', color='red', linewidth=2, label='Ideal Line')
plt.title('Linear Regression - Testing Set')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.legend()
plt.show()

# Print model evaluation metrics
print("The model performance for training set")
print("--------------------------------------")
print('RMSE is {:.2f}'.format(np.sqrt(mean_squared_error(Y_train, y_train_predict))))
print('R2 score is {:.2f}'.format(r2_score(Y_train, y_train_predict)))
print("\n")

print("The model performance for testing set")
print("--------------------------------------")
print('RMSE is {:.2f}'.format(np.sqrt(mean_squared_error(Y_test, y_test_predict))))
print('R2 score is {:.2f}'.format(r2_score(Y_test, y_test_predict)))
```
